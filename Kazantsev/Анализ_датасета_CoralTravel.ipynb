```python
import pandas as pd
import re

def extract_voucher_numbers(df, column_name):
    # Проверяем тип данных указанного столбца
    if df[column_name].dtype != object:
        print("Данные в указанном столбце не являются текстом.")
        return df

    # Извлекаем значения из указанного столбца с помощью регулярного выражения
    voucher_numbers = df[column_name].str.extract(r'\b(3\d{6}|4\d{6})\b', expand=False)
    # voucher_numbers = df[column_name].str.extract(r'(\b\d{7,8}\b)', expand=False)

    # Создаем новый столбец с полученными значениями
    new_column_name = column_name + "_voucher"
    df[new_column_name] = voucher_numbers

    return df
```

```python
# Вызываем функцию для извлечения номеров ваучеров для столбца 'Description'
df_work = extract_voucher_numbers(df_work, 'Description')

# Выводим обновленный датафрейм
df_work.head()
```

```python
# https://habr.com/ru/companies/otus/articles/755772/

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Пример текстовых данных

random_indices = np.random.choice(df_work['Description'].index, 1000, replace=False)  # Выбираем 1000 случайных индексов
documents = df_work.loc[random_indices, 'Description'].values  # Получаем значения по случайным индексам и сохраняем их в переменной 'documents'

# Создание объекта TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Применение TF-IDF к текстовым данным
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Определение порога для отбора ключевых слов
threshold = 0.2
important_words = [word for word, score in zip(tfidf_vectorizer.get_feature_names_out(), tfidf_matrix.toarray()[0]) if score >= threshold]
print("Важные слова:", important_words)
```
